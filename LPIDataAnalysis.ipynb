```python
usa_data = data[data['Economy'] == 'United States']
usa_data
```

```python
import matplotlib.pyplot as plt

usa_scores = usa_data[['Customs Score', 'Infrastructure Score', 'International Shipments Score',
                       'Logistics Competence and Quality Score', 'Timeliness Score', 'Tracking and Tracing Score']].values.flatten()

labels = ['Customs', 'Infrastructure', 'International Shipments', 'Logistics Competence', 'Timeliness', 'Tracking & Tracing']

plt.figure(figsize=(10, 6))
plt.bar(labels, usa_scores, color='skyblue')
plt.xlabel("Logistics Dimensions")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.title("Logistics Performance Scores for the United States")
plt.ylim(0, 5)
plt.show()
```

```python
import seaborn as sns

# Filter data for the U.S. and similar economies (e.g., Germany, Netherlands, Japan)
countries = ['United States', 'Germany', 'Netherlands', 'Japan']
filtered_data = data[data['Economy'].isin(countries)]
```

```python
# Define metrics of interest
metrics = ['Customs Score', 'Infrastructure Score', 'Timeliness Score']

# Plot a bar chart for each metric
for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.barplot(data=filtered_data, x='Economy', y=metric)
    plt.title(f"{metric} Comparison Across Economies")
    plt.ylabel(metric)
    plt.show()
```

```python
# Calculate average scores of other economies (excluding the U.S.) for 2023
avg_other_countries = filtered_data[filtered_data['Economy'] != 'United States'][metrics].mean()

# Get the U.S. scores for comparison
us_scores = filtered_data[filtered_data['Economy'] == 'United States'][metrics].iloc[0]

# Calculate the differences
score_differences = us_scores - avg_other_countries

# Display results
print("Score Differences Between U.S. and Other Top Economies (2023):")
print(score_differences)

# Visualize the differences with bar plots
plt.figure(figsize=(10, 6))
sns.barplot(x=score_differences.index, y=score_differences.values, palette="coolwarm")
plt.axhline(0, color='gray', linestyle='--')
plt.title("U.S. Score Differences from Average of Top Economies (2023)")
plt.ylabel("Difference in Score (U.S. - Average)")
plt.xlabel("Logistics Performance Dimensions")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load LPI data
# Assuming 'lpi_data' is a DataFrame that contains 'Country', 'LPI Score', and 'Group Rank' columns

# Prepare data for clustering
X = data[['LPI Score', 'LPI Grouped Rank']]

# Standardize the data for accurate clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
data['Cluster'] = kmeans.fit_predict(X_scaled)

# Map clusters to descriptive names based on cluster centers
# Identifying cluster labels based on relative LPI Scores and Group Rank averages
cluster_centers = kmeans.cluster_centers_
cluster_labels = cluster_centers[:, 0].argsort()  # Sort clusters by LPI Score

# Assign meaningful names based on scores (0: Best, 1: Average, 2: Worst)
cluster_names = {cluster_labels[0]: 'Best Trading Routes', 
                 cluster_labels[1]: 'Average Trading Routes', 
                 cluster_labels[2]: 'Worst Trading Routes'}

# Map cluster labels to categories
data['Cluster Category'] = data['Cluster'].map(cluster_names)

# Group countries by category
worst_routes = data[data['Cluster Category'] == 'Best Trading Routes']
average_routes = data[data['Cluster Category'] == 'Average Trading Routes']
best_routes = data[data['Cluster Category'] == 'Worst Trading Routes']

# Print the countries in each category
print("Countries with Best Trading Routes:")
display(best_routes[['Economy', 'LPI Score', 'LPI Grouped Rank']].head(20))

print("\nCountries with Average Trading Routes:")
display(average_routes[['Economy', 'LPI Score', 'LPI Grouped Rank']].head(20))

print("\nCountries with Worst Trading Routes:")
display(worst_routes[['Economy', 'LPI Score', 'LPI Grouped Rank']].head(20))
```
